```python
%%writefile diabeticapp.py

# importing libraries

import streamlit as st
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import seaborn as sns


#load dataset
df=pd.read_csv('/content/diabetes.csv')

#displying first 5 rows
df.head()

 #checking for null values
df.isnull().sum()

#checking how many type of values present
df['Outcome'].value_counts()

#grouping the data
df.groupby('Outcome').mean()
#0--> non_diabetic patient
#1--> diabetic patient

#Data visualization
df.groupby('Outcome')["Pregnancies"].mean().plot(kind='bar')

#  Create a pivot table
df.pivot_table(
    columns='Pregnancies',
    index='Outcome',
    values='Age',
    aggfunc='mean',
    fill_value=0 # Use an aggregation function (e.g., 'sum', 'count', 'mean', etc.)
)

# dividing data input and output variables
x=df.drop(columns='Outcome',axis=1) # input

y=df['Outcome'] #output

x.head()

y.head()

#import train and test spit
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=9,stratify=y) # dividing data into train and test sets

print(x_train.shape,x_test.shape,y_train.shape,y_test.shape) # checking the shape of array

# IMPORTING SVM CLASSIFIER
from sklearn.svm import SVC
model=SVC(kernel='linear') #linear(0,1)
model.fit(x_train,y_train)

#PREDICTING THE VALUES
y_predict=model.predict(x_test)
y_predict

# FINDING ACCURACY
accuracy=accuracy_score(y_test,y_predict)
accuracy



# EVALUATION
array = np.array( [[5, 17, 19, 32, 87, 54, 67, 25]])

prediction_array=model.predict(array)
if prediction_array[0]==0:
  print("Non-Diabetic")
else:
  print("Diabetic")

st.title("Diabetes Prediction for no.of pregnancies")
st.write("enter patient details to predict diabetes")

#getting input from user
pregnancies=st.number_input("No.of Pregnancies")
glucose=st.number_input("Glucose Level")
bloodpressure=st.number_input("Blood Pressure Value")
skinthickness=st.number_input("Skin Thickness Value")
insulin=st.number_input("Insulin Level")
bmi=st.number_input("BMI Value")
dpf=st.number_input("Diabetes Pedigree Function Value")
age=st.number_input("Age")

diagnosis=" "
if st.button("diabetes test results"):
    diagnosis=model.predict([[pregnancies,glucose,bloodpressure,skinthickness,insulin,bmi,dpf,age]])
    if diagnosis[0] == 0:
        st.success("Non-Diabetic")
    else:
        st.error("Diabetic")



```
